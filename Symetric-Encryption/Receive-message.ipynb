```python
#Import the cryptog raphy library
import cryptography
#Use a symmet ric encryption method
from cryptography.fernet import Fernet

#Load the key   #file generate of Send-message.ipynb
with open("secret.key", "rb") as key_file:
  key= key_file.read ()

#Create a Fernet object
cipher = Fernet(key)
```

```python
# Type secret message
encrypted_message = input("Write Secret message encrypted: ").encode()

# Decrypt the message
message = cipher.decrypt(encrypted_message)

#Print the original message and the decrypted mes sage
print("Original message : ", message.decode ())
```
